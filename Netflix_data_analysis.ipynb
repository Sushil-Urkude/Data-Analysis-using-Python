### Question Part

**Question1: For 'House of Cards', what is the Show Id and Who is the Director of this show ?**

```python
# To find the answer we have two methods
# First with isin() function
df[df['Title'].isin(['House of Cards'])]
```

**Question2: In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.**

```python
# To check the datatype of Released Date column
df.dtypes
```

```python
# To change datatype of 'Release_Date' column from object to datetime
df['Date_N']= pd.to_datetime(df['Release_Date'])
df.head()
```

```python
df.dtypes
```

```python
# To count the number of TV Shows & Moviesin particular year
year=df['Date_N'].dt.year.value_counts()
year
```

```python
# To show this value in form of Bar Plot
year.plot(kind='bar',color='r')
```

**Answer2: We can conclude from the above bar graph that in year 2019, Netflix released most number of TV shows and Movies.**
